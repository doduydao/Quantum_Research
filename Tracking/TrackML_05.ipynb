```python
dataset = pd.read_csv('event000001000/event000001000-hits.csv')
dataset.shape
```

```python
dataset_1 = pd.read_csv('event000001000/event000001000-truth-wo-noise.csv')
dataset_1.head()
```

```python
hit_id_truth_wo_noise_list = dataset_1['hit_id'].tolist()
len(hit_id_truth_wo_noise_list)
```

```python
volume_id = 9
hits_volume = dataset[['volume_id']==volume_id]
hits_volume
```

```python
dataset_2 = hits_volume[hits_volume['hit_id'].isin(hit_id_truth_wo_noise_list)]
dataset_2
```

```python
sublayer_2 = list(set(dataset_2['z']))[1::4]
len(sublayer_2)
```

```python
dataset_3 = dataset_2[dataset_2['z'].isin(sublayer_2)]
dataset_3
```

```python
hits_id_volume_9 = dataset_3['hit_id'].tolist()
truth_particle_id = list(dataset_1[dataset_1['hit_id'].isin(hits_id_volume_9)]['particle_id'])
```

```python
merged_df = dataset_3.merge(dataset_1[['hit_id', 'particle_id']], on='hit_id', how='left')
merged_df
```

```python
layer_id_list =  list(set(merged_df['layer_id']))
print(layer_id_list)
```

```python
particle_by_layer = dict()
for l in layer_id_list:
    d =  merged_df[merged_df['layer_id']==l]
    particle_by_layer[l] =  list(set(d['particle_id']))
    # print(set(dataset_3[dataset_3['particle_id'].isin(seclect)]
for k, v in particle_by_layer.items():
    print(k, len(v))
```

```python
import random
keys = list(particle_by_layer.keys())
smallest =  list(particle_by_layer[14])
select_particle_id = set()

for i in smallest:
    do_select = 0
    for k in keys[:-1]:
        if i in particle_by_layer[k]:
            do_select+=1
    if do_select ==5:
        select_particle_id.add(i)
select_particle_id =  list(select_particle_id)
print(select_particle_id)
idx = set()
while len(idx) < 10:
    idx.add(random.randint(0, len(select_particle_id)-1))
len(idx)

rand_particles = []
for i in idx:
    rand_particles.append(select_particle_id[i])
len(rand_particles)
```

```python
dataset_5 = merged_df[merged_df['particle_id'].isin(rand_particles)]
dataset_5.head(30)
```

```python
final = dataset_5.drop_duplicates(subset=['particle_id', 'layer_id'], keep='first')
del final['particle_id']
final.shape
```

```python
final.to_csv("event000001000/sel/event000001000-hits-sel-vol_9_10_track.csv",index=False)
```
